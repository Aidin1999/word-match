```python
unique_category_names = data['CategoryName'].unique()

# Translate CategoryName
category_name_translation = {name: GoogleTranslator(source='fa', target='en').translate(name) for name in unique_category_names}
data['CategoryNameEn'] = data['CategoryName'].map(category_name_translation)
```

```python
# Function to extract English brand name based on the Farsi BrandName
def extract_english_name(row):
    brand_words = row['BrandName'].split()  # Split BrandName into words
    name_words = row['Name'].split()       # Split Name into words
    num_words = min(len(brand_words), 3)   # Use up to 3 words
    return " ".join(name_words[:num_words]).lower()  # Extract first num_words from Name in lowercase

# Apply the function to create a proper BrandNameEn column
data['BrandNameEn'] = data.apply(extract_english_name, axis=1)
```

```python
category_keys = list(data['CategoryName'].values) + list(data['CategoryNameEn'].values)

# Create tokenized categories dictionary
tokenized_categories = {
    category: ".".join(category.lower().replace(" and ", " ").replace("(", "").replace(")", "").split())
    for category in category_keys
}

# Function to tokenize categories
def get_tokenized_category(category):
    category = (
        category.lower()
        .replace(" and ", " ")  # Remove "and"
        .replace("(", "")       # Remove "("
        .replace(")", "")       # Remove ")"
    )
    return tokenized_categories.get(category, ".".join(category.split()))

# Apply the function to combine tokens from both columns
data['CategoryNameTokens'] = (
    data['CategoryNameEn'].apply(get_tokenized_category) + "." + data['CategoryName'].apply(get_tokenized_category)
)
```

```python
def clean_model_with_tokens(model, tokens, farsi_brand, english_brand, extra_words):
    # Normalize the model: lower case
    model = model.lower()
    # Remove brand names
    model = model.replace(farsi_brand.lower(), "").replace(english_brand.lower(), "").strip()
    # Remove extra words
    for word in extra_words:
        model = model.replace(word, "")
    # Remove category tokens
    token_list = tokens.split('.')
    for token in token_list:
        model = model.replace(token, "")
    # Return cleaned model
    return " ".join(model.split())  # Remove extra spaces

# Extra words to remove
extra_words_to_remove = ["مدل", "سری",'Galaxy','galaxy']

# Apply the cleaning function to FaModel and EnModel
data['FaModel'] = data.apply(
    lambda row: clean_model_with_tokens(
        row['FaModel'], row['CategoryNameTokens'], row['BrandName'], row['BrandNameEn'], extra_words_to_remove
    ),
    axis=1
)
data['EnModel'] = data.apply(
    lambda row: clean_model_with_tokens(
        row['EnModel'], row['CategoryNameTokens'], row['BrandName'], row['BrandNameEn'], extra_words_to_remove
    ),
    axis=1
)
```
